## **1.Cut video to frame**

```python
import cv2
import os
from datetime import datetime, timedelta


def video_to_frames(input_loc, output_loc, desired_fps=2, time_watermark=True,video_name="test1", video_datetime="1970-01-01 00:00:00.000"):
    """
    Extract frames from an input video file and save them as separate frames in an output directory,
    with datetime watermarks on each frame and filenames including the datetime.

    Args:
        input_loc: Input video file.
        output_loc: Output directory to save the frames.
        desired_fps: Desired frames per second for extraction.
        video_name: Prefix for output frame filenames.
        video_datetime: Datetime the video was taken in "YYYY-MM-DD HH:mm:00" format.
    """
    try:
        os.mkdir(output_loc)
    except OSError:
        pass

    # Start capturing the video feed
    cap = cv2.VideoCapture(input_loc)
    actual_fps = cap.get(cv2.CAP_PROP_FPS)

    # Validate and adjust the desired FPS
    if desired_fps > actual_fps or desired_fps <= 0:
        print(f"Desired FPS ({desired_fps}) is invalid. Using actual FPS ({actual_fps}).")
        desired_fps = actual_fps

    frame_interval = int(actual_fps / desired_fps)
    if frame_interval <= 0:
        frame_interval = 1

    # Parse video_datetime into a datetime object
    base_datetime = datetime.strptime(video_datetime, "%Y-%m-%d %H:%M:%S.%f")

    # Log video details
    video_length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"Video Length: {video_length} frames")
    print(f"Video taken at: {video_datetime}")
    print(f"Actual FPS: {actual_fps}")
    print(f"Desired FPS: {desired_fps}")
    print(f"Frame Interval: {frame_interval}")
    print("Converting video...\n")

    count = 0
    frame_count = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Save frames at the specified interval
        if count % frame_interval == 0:
            # Calculate the datetime for the current frame
            frame_time_delta = timedelta(seconds=frame_count / desired_fps)
            current_datetime = base_datetime + frame_time_delta
            timestamp_str = current_datetime.strftime("%Y-%m-%d %H:%M:%S.%f")

            # Add watermark with the original video datetime for the first frame
            font = cv2.FONT_HERSHEY_SIMPLEX
            font_scale = 1
            color = (0, 255, 0)  # Green
            thickness = 2
            position = (10, frame.shape[0] - 10)  # Bottom-left corner

            if(time_watermark):
                frame_watermark = f"Time: {current_datetime}"
                frame = cv2.putText(frame, frame_watermark, position, font, font_scale, color, thickness)
            

            # Save the frame with timestamp in filename
            frame_filename = os.path.join(
                output_loc, f"{video_name}_frame_{current_datetime.strftime('%Y_%m_%d_%H_%M_%S.%f')}.jpg"
            )
            cv2.imwrite(frame_filename, frame)
            print(f"Saved frame {frame_count} at {timestamp_str}")
            frame_count += 1

        count += 1

    # Release the video capture
    cap.release()
    print(f"Done extracting frames. {frame_count} frames saved.")

# Example Usage

```

## <u>**Draw the outline for labeling the frame**</u>

```python
import os
import json
import cv2
import numpy as np
from pathlib import Path
from tqdm import tqdm
from itertools import groupby

# Input paths
images_folder = "a360_photos"  # Folder containing the original images
coco_file = "labeled.json"     # Path to the COCO JSON file

# Output path
output_folder = "a360_labeled_images"
os.makedirs(output_folder, exist_ok=True)

# Category ID mapping with colors and positions
category_mapping = {
    1: {"label": "cargo_door", "color": ( 0, 0,255), "text_position": "left_up"},  # Red
    2: {"label": "cargo door open", "color": ( 47, 47,211), "text_position": "left_up"},  # Hot Pink
    3: {"label": "cargo door closed", "color": ( 115, 115,229), "text_position": "left_up"},  # Light Red
    4: {"label": "cargo door processing (open or close)", "color": (166, 143, 233), "text_position": "left_up"},  # Pink
    5: {"label": "JCPL", "color": (2, 82, 228), "text_position": "bottom_left"},  # Dark Orange
    6: {"label": "JCPL docked", "color": (3, 99, 223), "text_position": "bottom_left"},  # Orange
    7: {"label": "JCPL not docked", "color": (46, 185, 241), "text_position": "bottom_left"},  # Yellow
    8: {"label": "ULD", "color": (123, 40, 32), "text_position": "left_up"},  # Blue
    9: {"label": "ULD waiting", "color": (145, 30, 111), "text_position": "left_up"},  # Purple
    10: {"label": "ULD handling", "color": (156, 92, 51), "text_position": "left_up"},  # Dark Blue
}

# Load COCO annotations
with open(coco_file, 'r') as f:
    coco_data = json.load(f)

# Create a mapping from image ID to filename
image_id_to_filename = {
    img["id"]: os.path.basename(img["file_name"]) for img in coco_data["images"]
}

# Parse annotations
annotations = coco_data["annotations"]

# Process each image
for image_id, image_annotations in tqdm(
    groupby(sorted(annotations, key=lambda x: x["image_id"]), key=lambda x: x["image_id"]),
    desc="Processing images"
):
    image_annotations = list(image_annotations)  # Convert groupby object to list
    image_file = image_id_to_filename.get(image_id)
    if not image_file:
        continue

    image_path = os.path.join(images_folder, image_file)
    if not os.path.exists(image_path):
        print(f"Image file not found: {image_path}")
        continue

    # Read the image
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading image: {image_path}")
        continue

    # Process annotations for this image
    for annotation in image_annotations:
        segmentation = annotation.get("segmentation", [])
        category_id = annotation.get("category_id", "unknown")

        # Skip if category ID is invalid
        if category_id not in category_mapping:
            continue

        # Get label, color, and position
        category_info = category_mapping[category_id]
        label = category_info["label"]
        color = category_info["color"]
        text_position = category_info["text_position"]

        # Draw segmentation
        for segment in segmentation:
            points = np.array(segment).reshape(-1, 2).astype(int)
            cv2.polylines(image, [points], isClosed=True, color=color, thickness=2)

            # Add label
            if points.size > 0:
                if text_position == "left_up":
                    x, y = points.min(axis=0)  # Top-left corner
                    text_coords = (x, y - 5)
                elif text_position == "bottom_left":
                    x = points.min(axis=0)[0]  # Minimum x (leftmost point)
                    y = points.max(axis=0)[1]  # Maximum y (bottommost point)
                    text_coords = (x, y + 15)  # Offset slightly below the bottom

                cv2.putText(
                    image,
                    label,
                    text_coords,
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1,
                    color,
                    2,
                    cv2.LINE_AA
                )

    # Save the annotated image
    output_path = os.path.join(output_folder, image_file)
    cv2.imwrite(output_path, image)

print(f"Annotated images saved to: {output_folder}")

```

```python
import os
import cv2
from datetime import datetime, timedelta

# Input Milestones
milestone = {
    'cargo_door_open': datetime(2024, 11, 22, 13, 39, 30),
    'jcpl_docked': datetime(2024, 11, 22, 13, 39, 31, 500000),
    'first_uld_handling': datetime(2024, 11, 22, 13, 39, 56, 500000),
    'jcpl_undocked': datetime(2024, 11, 22, 13, 46, 48),
    'last_uld_handling': datetime(2024, 11, 22, 13, 45, 52),
    'cargo_door_closed': datetime(2024, 11, 22, 13, 48, 47)
}

# Parameters
frame_folder = "a360_labeled_images"  # Folder containing labeled frames
video_name = "IMG_1937"  # Base name for the video
label_duration = timedelta(seconds=120)  # Duration for the labels to be displayed
final_frame_folder = "a360_final_frames"  # Folder containing labeled frames

# Helper Functions
def get_timestamp_from_filename(filename):
    """
    Extracts the timestamp from the frame filename.

    :param filename: Frame filename with timestamp.
    :return: Datetime object extracted from the filename.
    """
    timestamp_str = filename.split(f"{video_name}_frame_")[1]
    timestamp_str = timestamp_str.rsplit(".", 1)[0]  # Remove extension
    return datetime.strptime(timestamp_str, "%Y_%m_%d_%H_%M_%S.%f")  # Convert string to datetime object

def draw_labels_on_frame(image, labels):
    """
    Draws multiple labels vertically aligned in the top-left corner of the image.

    :param image: Image (numpy array).
    :param labels: List of labels to display.
    :return: Image with the labels drawn.
    """
    # Box and text properties
    font = cv2.FONT_HERSHEY_TRIPLEX
    font_scale = 1.0  # Increased font size
    font_thickness = 2  # Thicker text
    outline_color = (128, 0, 128)  # Purple outline color
    fill_color = (200, 162, 200)  # Light purple fill color
    padding = 20  # Padding from the corner
    text_padding = 15  # Padding inside the box around the text
    vertical_spacing = 10  # Space between consecutive labels

    # Initialize y-coordinate for the first label
    current_y = padding

    for label_text in labels:
        # Get text size
        text_size = cv2.getTextSize(label_text, font, font_scale, font_thickness)[0]
        text_width, text_height = text_size

        # Define box coordinates
        box_start = (padding, current_y)
        box_end = (
            padding + text_width + 2 * text_padding,
            current_y + text_height + 2 * text_padding,
        )

        # Draw filled rectangle for the background
        cv2.rectangle(image, box_start, box_end, fill_color, -1)

        # Draw the outline rectangle
        cv2.rectangle(image, box_start, box_end, outline_color, 2)

        # Calculate text position
        text_x = box_start[0] + text_padding
        text_y = current_y + text_height + text_padding

        # Draw the label text
        cv2.putText(image, label_text, (text_x, text_y), font, font_scale, outline_color, font_thickness)

        # Update y-coordinate for the next label
        current_y = box_end[1] + vertical_spacing

    return image


# Process Frames
for frame_file in frames:
    # Get frame timestamp
    frame_path = os.path.join(frame_folder, frame_file)
    frame_timestamp = get_timestamp_from_filename(frame_file)

    # Collect labels for the current frame
    labels_to_draw = []
    for milestone_name, milestone_time in milestone.items():
        if milestone_time <= frame_timestamp <= milestone_time + label_duration:
            labels_to_draw.append(milestone_name.replace("_", " ").capitalize())

    # Add labels if applicable
    if labels_to_draw:
        image = cv2.imread(frame_path)
        image_with_labels = draw_labels_on_frame(image, labels_to_draw)

        # Save the labeled frame
        output_path = os.path.join(final_frame_folder, frame_file)
        os.makedirs(final_frame_folder, exist_ok=True)
        cv2.imwrite(output_path, image_with_labels)
        print(f"Labeled frame saved to: {output_path}")

```

```python
import cv2
import os
from datetime import datetime

# Parameters
frame_folder = "a360_final_frames"  # Folder containing labeled frames
video_name = "IMG_1937"  # Base name for the video
output_video = video_name+"15xfast_labeled.mp4" # Output video file name
fps = 30  # Frames per second



# Get all frame filenames and sort by timestamp in the filename
frames = [
    f for f in os.listdir(frame_folder)
    if f.startswith(f"{video_name}_frame_") and f.endswith(".jpg")  # Adjust extension if needed
]

# Sort frames by timestamp in the filename
def get_timestamp_from_filename(filename):
    # Extract timestamp portion from the filename by splitting only at the last dot
    timestamp_str = filename.split(f"{video_name}_frame_")[1]
    timestamp_str = timestamp_str.rsplit(".", 1)[0]  # Split at the last dot to remove the extension
    print(timestamp_str)
    return datetime.strptime(timestamp_str, "%Y_%m_%d_%H_%M_%S.%f")  # Convert string to datetime object

# Sort frames by the timestamp
frames.sort(key=get_timestamp_from_filename)

# Ensure there are frames to process
if not frames:
    print("No frames found in the folder.")
    exit()

# Get frame size from the first frame
first_frame_path = os.path.join(frame_folder, frames[0])
frame = cv2.imread(first_frame_path)
if frame is None:
    print(f"Could not read the first frame: {first_frame_path}")
    exit()

height, width, _ = frame.shape

# Define the video writer
fourcc = cv2.VideoWriter_fourcc(*"mp4v")  # Codec for .mp4 format
video_writer = cv2.VideoWriter(output_video, fourcc, fps, (width, height))

# Add frames to the video
for frame_name in frames:
    frame_path = os.path.join(frame_folder, frame_name)
    frame = cv2.imread(frame_path)
    if frame is None:
        print(f"Could not read frame: {frame_path}")
        continue
    video_writer.write(frame)

# Release the video writer
video_writer.release()
print(f"Video created successfully: {output_video}")
```

```python
import os
import json
from datetime import datetime
import pandas as pd

# Step 1: Process the labeled JSON and create a table

def parse_labeled_json(json_path, category_mapping):
    # Load JSON file
    with open(json_path, 'r') as file:
        data = json.load(file)

    # Prepare rows for the table
    rows = []

    for image in data.get("images", []):
        # Extract file name and parse datetime
        file_name = os.path.basename(image["file_name"])
        try:
            timestamp_str = file_name.split("_frame_")[1].rsplit(".", 1)[0]
            timestamp = datetime.strptime(timestamp_str, "%Y_%m_%d_%H_%M_%S.%f")
        except (IndexError, ValueError):
            timestamp = None  # Set to null if parsing fails

        # Find annotations for the image
        image_id = image["id"]
        annotations = [ann for ann in data.get("annotations", []) if ann["image_id"] == image_id]

        # Add rows for each annotation
        for ann in annotations:
            category_id = ann["category_id"]
            category_label = category_mapping.get(category_id, {}).get("label", "Unknown")
            rows.append({
                "image_name": file_name,
                "timestamp": timestamp,
                "category_id": category_id,
                "category_label": category_label
            })

    # Convert rows to a DataFrame
    df = pd.DataFrame(rows)

    # Add a column for the number of labels per image
    label_counts = df.groupby("image_name").size().reset_index(name="num_labels")
    df = df.merge(label_counts, on="image_name", how="left")

    return df

# Category mapping as provided
category_mapping = {
    1: {"label": "cargo_door", "color": (0, 0, 255), "text_position": "left_up"},
    2: {"label": "cargo door open", "color": (47, 47, 211), "text_position": "left_up"},
    3: {"label": "cargo door closed", "color": (115, 115, 229), "text_position": "left_up"},
    4: {"label": "cargo door processing (open or close)", "color": (166, 143, 233), "text_position": "left_up"},
    5: {"label": "JCPL", "color": (2, 82, 228), "text_position": "bottom_left"},
    6: {"label": "JCPL docked", "color": (3, 99, 223), "text_position": "bottom_left"},
    7: {"label": "JCPL not docked", "color": (46, 185, 241), "text_position": "bottom_left"},
    8: {"label": "ULD", "color": (123, 40, 32), "text_position": "left_up"},
    9: {"label": "ULD waiting", "color": (145, 30, 111), "text_position": "left_up"},
    10: {"label": "ULD handling", "color": (156, 92, 51), "text_position": "left_up"},
}

# Example usage
json_path = "labeled.json"  # Path to your Azure ML labeling file
df = parse_labeled_json(json_path, category_mapping)
df.head(10)
```

```python
# Step 2: Detect milestones in the video

def detect_milestones(df, video_name):
    milestones = {}

    # Filter by the specific video
    df = df[df["image_name"].str.startswith(video_name)]

    # Sort by timestamp
    df = df.sort_values("timestamp").reset_index(drop=True)

    # Detect cargo door open (category_id=2)
    cargo_door_open = df[df["category_id"] == 2].iloc[0]
    milestones["cargo_door_open"] = cargo_door_open["timestamp"]

    # Detect JCPL docked (category_id=6) after cargo door open
    jcpl_docked = df[(df["category_id"] == 6) & (df["timestamp"] > milestones["cargo_door_open"])].iloc[0]
    milestones["jcpl_docked"] = jcpl_docked["timestamp"]

    # Detect first ULD handling (category_id=10) after JCPL docked
    uld_handling_start = df[(df["category_id"] == 10) & (df["timestamp"] > milestones["jcpl_docked"])].iloc[0]
    milestones["uld_handling_start"] = uld_handling_start["timestamp"]

    # Detect JCPL not docked (category_id=7) after first ULD handling
    jcpl_not_docked = df[(df["category_id"] == 7) & (df["timestamp"] > milestones["uld_handling_start"])].iloc[0]
    milestones["jcpl_not_docked"] = jcpl_not_docked["timestamp"]

    # Detect last ULD handling (category_id=10) before JCPL not docked
    last_uld_handling = df[(df["category_id"] == 10) & (df["timestamp"] < milestones["jcpl_not_docked"])].iloc[-1]
    milestones["last_uld_handling"] = last_uld_handling["timestamp"]

    # Detect cargo door closed (category_id=3) after JCPL not docked
    cargo_door_closed = df[(df["category_id"] == 3) & (df["timestamp"] > milestones["jcpl_not_docked"])].iloc[0]
    milestones["cargo_door_closed"] = cargo_door_closed["timestamp"]

    # Calculate durations
    milestones["cargo_door_open_duration"] = (
        milestones["cargo_door_closed"] - milestones["cargo_door_open"]
    ).total_seconds()

    return milestones

# Example usage
video_name = "IMG_1937"
milestones = detect_milestones(df, video_name)
print(milestones)
```

## convert the coco file to YOLO type

```python
import json
import os

def convert_coco_to_yolo_segmentation(coco_file_path, output_dir, image_dir):
    """
    Converts Azure ML COCO segmentation output to YOLO segmentation format.

    :param coco_file_path: Path to the COCO JSON file.
    :param output_dir: Directory to save YOLO format files.
    :param image_dir: Directory containing images referenced in the COCO JSON file.
    """
    with open(coco_file_path, 'r') as f:
        coco_data = json.load(f)

    # Map category IDs to YOLO class indices
    category_map = {cat['id']: i for i, cat in enumerate(coco_data['categories'])}

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Iterate over all annotations
    for annotation in coco_data['annotations']:
        image_id = annotation['image_id']
        category_id = annotation['category_id']
        segmentation = annotation['segmentation']  # Segmentation points (list of polygons)

        # Get image details to normalize coordinates
        image_info = next(img for img in coco_data['images'] if img['id'] == image_id)
        img_width = image_info['width']
        img_height = image_info['height']

        # Extract the file name (handle potential directory structure)
        full_image_name = image_info['file_name']
        image_name = os.path.basename(full_image_name)  # Gets the string after the last "/"

        # Prepare YOLO format file for the image
        yolo_file_name = os.path.splitext(image_name)[0] + ".txt"
        yolo_file_path = os.path.join(output_dir, yolo_file_name)

        with open(yolo_file_path, 'a') as yolo_file:
            # Loop over each polygon in the segmentation
            for polygon in segmentation:
                # Normalize the polygon coordinates
                normalized_points = [
                    (x / img_width, y / img_height)
                    for x, y in zip(polygon[::2], polygon[1::2])
                ]

                # Convert points to YOLO format string
                normalized_points_str = " ".join(
                    f"{x:.6f} {y:.6f}" for x, y in normalized_points
                )

                # Write the category ID and normalized points
                yolo_file.write(f"{category_map[category_id]} {normalized_points_str}\n")

    print(f"YOLO segmentation files saved in {output_dir}")


# Example usage
coco_file_path = "labeled.json"
output_dir = "yolo/yolo_output"
image_dir = "a360_photos"

convert_coco_to_yolo_segmentation(coco_file_path, output_dir, image_dir)
```

## Split and train the Yolo model 

```python
import os
import random
import shutil

def split_data(output_dir, image_dir, train_ratio=0.7, val_ratio=0.2, dataset_dir="dataset"):
    """
    Splits YOLO dataset into training, validation, and test sets, ensuring images without labels are removed.

    :param output_dir: Path to the YOLO annotations directory (contains labels/).
    :param image_dir: Path to the directory containing images.
    :param train_ratio: Ratio of training data (default is 70% training).
    :param val_ratio: Ratio of validation data (default is 20% validation).
    :param dataset_dir: Directory to save the split dataset.
    """
    labels_dir = output_dir

    # Create train/val/test directories
    for split in ["train", "val", "test"]:
        os.makedirs(os.path.join(dataset_dir, "images", split), exist_ok=True)
        os.makedirs(os.path.join(dataset_dir, "labels", split), exist_ok=True)

    # Pair images with their corresponding labels and remove unlabeled images
    all_data = []
    for image_file in os.listdir(image_dir):
        if image_file.endswith(('.jpg', '.png')):
            # Extract the base name (everything before the last dot)
            base_name = ".".join(image_file.split('.')[:-1])
            label_file = f"{base_name}.txt"
            label_path = os.path.join(labels_dir, label_file)
            if os.path.exists(label_path):
                all_data.append((image_file, label_file))
            else:
                # Remove image without a corresponding label
                image_path = os.path.join(image_dir, image_file)
                os.remove(image_path)
                print(f"Removed unlabeled image: {image_path}")

    # Shuffle and split data
    random.shuffle(all_data)
    total_size = len(all_data)
    train_size = int(total_size * train_ratio)
    val_size = int(total_size * val_ratio)
    test_size = total_size - train_size - val_size

    train_data = all_data[:train_size]
    val_data = all_data[train_size:train_size + val_size]
    test_data = all_data[train_size + val_size:]

    # Move files to respective directories
    for split, split_data in [("train", train_data), ("val", val_data), ("test", test_data)]:
        for image_file, label_file in split_data:
            # Copy image
            src_image_path = os.path.join(image_dir, image_file)
            dst_image_path = os.path.join(dataset_dir, "images", split, image_file)
            shutil.copy(src_image_path, dst_image_path)

            # Copy corresponding label
            src_label_path = os.path.join(labels_dir, label_file)
            dst_label_path = os.path.join(dataset_dir, "labels", split, label_file)
            shutil.copy(src_label_path, dst_label_path)

    print(f"Dataset split completed. Saved in '{dataset_dir}'.")
    print(f"Train size: {train_size}, Validation size: {val_size}, Test size: {test_size}")

# Paths
output_dir = "yolo/yolo_output"
image_dir = "a360_photos"
dataset_dir = "dataset"

# Split dataset
split_data(output_dir, image_dir, train_ratio=0.7, val_ratio=0.2, dataset_dir=dataset_dir)





```

```python
from ultralytics import YOLO
import os

def train_yolov8(dataset_dir, model_name="yolov8s", epochs=50, img_size=640):
    """
    Trains a YOLOv8 model on the dataset containing polygons as annotations.

    :param dataset_dir: Path to the dataset directory containing images/ and labels/.
    :param model_name: YOLOv8 model variant (e.g., yolov8s, yolov8m).
    :param epochs: Number of training epochs.
    :param img_size: Training image size.
    """
    # Category mapping for reference
    category_mapping = {
        1: {"label": "cargo_door", "color": (0, 0, 255), "text_position": "left_up"},
        2: {"label": "cargo door open", "color": (47, 47, 211), "text_position": "left_up"},
        3: {"label": "cargo door closed", "color": (115, 115, 229), "text_position": "left_up"},
        4: {"label": "cargo door processing (open or close)", "color": (166, 143, 233), "text_position": "left_up"},
        5: {"label": "JCPL", "color": (2, 82, 228), "text_position": "bottom_left"},
        6: {"label": "JCPL docked", "color": (3, 99, 223), "text_position": "bottom_left"},
        7: {"label": "JCPL not docked", "color": (46, 185, 241), "text_position": "bottom_left"},
        8: {"label": "ULD", "color": (123, 40, 32), "text_position": "left_up"},
        9: {"label": "ULD waiting", "color": (145, 30, 111), "text_position": "left_up"},
        10: {"label": "ULD handling", "color": (156, 92, 51), "text_position": "left_up"},
    }

    # Create the dataset.yaml content
    labels = [v["label"] for v in category_mapping.values()]
    dataset_abs_dir = os.path.abspath(dataset_dir)
    dataset_yaml = f"""path: {dataset_abs_dir}
train: images/train
val: images/val
test: images/test

nc: {len(labels)}
names:
  - cargo_door
  - cargo door open
  - cargo door closed
  - cargo door processing (open or close)
  - JCPL
  - JCPL docked
  - JCPL not docked
  - ULD
  - ULD waiting
  - ULD handling
"""

    # Save dataset.yaml
    dataset_yaml_path = os.path.join(dataset_dir, "dataset.yaml")
    with open(dataset_yaml_path, "w") as f:
        f.write(dataset_yaml)

    print("dataset.yaml generated successfully!")

    # Load the YOLOv8 model
    model = YOLO(f"{model_name}.pt")  # Ensure the pretrained weights are in the current directory or specify the correct path
    
    # If your model version supports polygon segmentation, you can directly use polygons for training:
    model.train(data=dataset_yaml_path, epochs=epochs, imgsz=img_size, batch=16, task='segment')

    print("Training completed.")
    return model


# Train YOLOv8
model = train_yolov8("dataset", model_name="yolov8m-seg", epochs=50, img_size=640)


```

```python
import matplotlib.pyplot as plt
import pandas as pd
def evaluate_yolov8(model_name="yolov8s", dataset_dir="dataset", split="val"):
    """
    Evaluates the YOLOv8 model on the specified dataset split.

    :param model_name: YOLOv8 model variant or path to trained model weights.
    :param dataset_dir: Path to the dataset directory containing dataset.yaml.
    :param split: Dataset split to evaluate (e.g., "val" or "test").
    """
    dataset_yaml_path = os.path.join(dataset_dir, "dataset.yaml")
    model = YOLO(f"{model_name}.pt")  # Ensure the correct model weights path
    results = model.val(data=dataset_yaml_path, split=split)  # Specify the split ('val' or 'test')

    return results

def summarize_results(results):
    # Extract relevant metrics from the results object
    mAP50 = results.box.map50  # mAP at 50% IoU
    mAP50_95 = results.box.map  # mAP at 50-95% IoU
    print(f"mAP@50: {mAP50:.4f}")
    print(f"mAP@50-95: {mAP50_95:.4f}")
    print(f"Precision: {results.box.precision.mean():.4f}")
    print(f"Recall: {results.box.recall.mean():.4f}")

def plot_pr_curve(results):
    precision, recall = results.box.pr  # Get precision-recall arrays
    plt.figure(figsize=(8, 6))
    plt.plot(recall, precision, label='Precision-Recall Curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.grid()
    plt.show()

def plot_confidence_curves(results):
    confidences = results.box.confidence
    precision = results.box.precision.mean(axis=1)
    recall = results.box.recall.mean(axis=1)

    plt.figure(figsize=(8, 6))
    plt.plot(confidences, precision, label='Precision')
    plt.plot(confidences, recall, label='Recall')
    plt.xlabel('Confidence Threshold')
    plt.ylabel('Metric Value')
    plt.title('Metrics vs Confidence Threshold')
    plt.legend()
    plt.grid()
    plt.show()

def class_wise_metrics(results):
    data = {
        "Class": results.ap_class_index,
        "Precision": results.box.precision,
        "Recall": results.box.recall,
        "AP": results.box.ap
    }
    df = pd.DataFrame(data)
    print(df)
    return df




# Evaluate YOLOv8
eva_results=evaluate_yolov8(model_name="yolov8s", dataset_dir="dataset", split="val")
summarize_results(eva_results)
# sum_df=class_wise_metrics(eva_results)
plot_pr_curve(eva_results)
plot_confidence_curves(eva_results)


```
