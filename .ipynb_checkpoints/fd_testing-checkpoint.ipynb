```python
output_label_dir='PD_CT_usecase/yolo_labels'
manifest_path=local_path
import os
os.makedirs(output_label_dir, exist_ok=True)
with open(manifest_path, 'r') as f:
    for line in f:
        item = json.loads(line)
        
        # 1. Extract image file name
        image_uri = item['source-ref']
        image_filename = os.path.basename(image_uri)
        image_name_no_ext = os.path.splitext(image_filename)[0]

        # 2. Get job name key (e.g., "LabelingJob-XYZ-metadata")
        metadata_key = next(k for k in item if k.endswith('-metadata') and 'job-name' in item[k])
        job_name = item[metadata_key]['job-name'].split('/')[-1]  # e.g., "Your-Job-Name"

        # 3. Find the actual annotation key that matches job_name (case-insensitive)
        annotation_key = next((k for k in item if k.lower() == job_name.lower()), None)
        if annotation_key is None or 'annotations' not in item[annotation_key]:
            
            continue

        # 4. Extract image size
        annotations = item[annotation_key]['annotations']
        image_size = item[annotation_key]['image_size'][0]
        img_w, img_h = image_size['width'], image_size['height']

        # 5. Convert to YOLO format
        yolo_lines = []
        for ann in annotations:
            class_id = ann['class_id']
            x_center = (ann['left'] + ann['width'] / 2) / img_w
            y_center = (ann['top'] + ann['height'] / 2) / img_h
            w_norm = ann['width'] / img_w
            h_norm = ann['height'] / img_h
            yolo_lines.append(f"{class_id} {x_center:.6f} {y_center:.6f} {w_norm:.6f} {h_norm:.6f}")

        # 6. Save .txt label file
        with open(os.path.join(output_label_dir, f"{image_name_no_ext}.txt"), 'w') as out_f:
            out_f.write('\n'.join(yolo_lines))
```

```python
import os
import matplotlib.pyplot as plt
from sagemaker.s3 import S3Downloader
from sagemaker import get_execution_role, Session

def show_conf_matrix_from_s3(bucket_name, image_directory, image_name='conf_matrix.png', local_download_dir='/tmp/cm_download'):
    s3_uri = f's3://{bucket_name}/{image_directory}/{image_name}'
    os.makedirs(local_download_dir, exist_ok=True)
    local_path = os.path.join(local_download_dir, image_name)

    # Create SageMaker session using current execution role
    role = get_execution_role()
    sm_session = Session()
    print(f"Using SageMaker execution role: {role}")
    print(f"Downloading from: {s3_uri}")

    # Download the image
    S3Downloader.download(s3_uri, local_path, sagemaker_session=sm_session)

    # Display
    img = plt.imread(local_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title("Confusion Matrix from S3")
    plt.show()

show_conf_matrix_from_s3(
    bucket_name='sagemaker-ap-southeast-1-302263047425',
    image_directory='processing-output/yolo-model-2025-06-30-025416/exp',
    image_name='confusion_matrix.png',
    local_download_dir='yolo_output'
)
```

```python
from datetime import datetime
import time
from sagemaker.sklearn.processing import SKLearnProcessor
from sagemaker.processing import ProcessingInput, ProcessingOutput

# Parameters
threshold = "5"  # Set hash difference threshold
input_bucket = "test-sats-label-data"
input_prefix = "stage=raw/test-02"  # Folder containing cut frames
output_bucket = "sagemaker-ap-southeast-1-302263047425"
output_prefix = "processed_frame"

# Generate unique job name and output path
current_datetime = datetime.now().strftime('%Y-%m-%d-%H%M%S')
input_data = f"s3://{input_bucket}/{input_prefix}"
output_path = f"s3://{output_bucket}/{output_prefix}"

print("Starting processing job for cut frames folder...")
start_time = time.time()

# Create processor with only threshold in env
sklearn_processor = SKLearnProcessor(
    framework_version="0.20.0",
    role=role,
    instance_type="ml.m5.xlarge",
    instance_count=1,
    env={
        "THRESHOLD": threshold
    }
)

# Run SageMaker processing job
sklearn_processor.run(
    code=script_s3_uri,
    inputs=[
        ProcessingInput(
            source=input_data,
            destination="/opt/ml/processing/input/frames"
        ),
        ProcessingInput(
            source=requirements_s3_uri,  # <-- NEW: requirements.txt input
            destination="/opt/ml/processing/input/requirements"
        )
    ],
    
    outputs=[
        ProcessingOutput(
            output_name="filtered_frames",
            source="/opt/ml/processing/output",
            destination=output_path
        )
    ]
)

end_time = time.time()
print(f"Processing job completed in {end_time - start_time:.2f} seconds")

# Retrieve and print output S3 URI
processing_job_description = sklearn_processor.jobs[-1].describe()
output_config = processing_job_description["ProcessingOutputConfig"]
for output in output_config["Outputs"]:
    if output["OutputName"] == "filtered_frames":
        print(f"Filtered output S3 URI: {output['S3Output']['S3Uri']}")

```
